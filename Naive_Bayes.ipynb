```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("/Users/bhavyakansal/AI/Datasets/Placement2.csv")
```

```python
dataset.head(10)
```

```python
plt.figure(figsize=(3,2))
sns.scatterplot(x="Cgpa" , y ="Score" , data=dataset , hue="Placed")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test , y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB , MultinomialNB , BernoulliNB
```

# GaussianNB

```python
gnb = GaussianNB()
gnb.fit(x_train , y_train)
```

```python
gnb.score(x_test,y_test)*100 , gnb.score(x_train,y_train)*100
```

```python
gnb.predict([[6.8,123]])
```

# Multinomial

```python
mnb = MultinomialNB()
mnb.fit(x_train , y_train)
```

```python
mnb.score(x_test,y_test)*100 , mnb.score(x_train,y_train)*100
```

```python
mnb.predict([[6.8,123]])
```

# BernoulliNB

```python
bnb = MultinomialNB()
bnb.fit(x_train , y_train)
```

```python
bnb.score(x_test,y_test)*100 , bnb.score(x_train,y_train)*100
```

```python
bnb.predict([[6.8,123]])
```

# Decision Region

```python
# Gaussian NB
```

```python
plt.figure(figsize=(3,3))
plot_decision_regions(x.to_numpy() , y.to_numpy(), clf=gnb)
```

```python
# Multinomial NB
```

```python
plt.figure(figsize=(3,3))
plot_decision_regions(x.to_numpy() , y.to_numpy(), clf=mnb)
```

```python
# Bernoulli NB
```

```python
plt.figure(figsize=(3,3))
plot_decision_regions(x.to_numpy() , y.to_numpy(), clf=bnb)
```
