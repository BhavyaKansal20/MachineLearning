**ODINAL ENCODING**

```python
#New Dataset
new_df = pd.DataFrame({"Size": ["s", "m", "l","xl", "s" , "m", "l","s", "s", "l", "xl", "m"]})
new_df.head(3)
```

```python
ord_data = [["s" , "m" , "l" , "xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(new_df[["Size"]] )
```

```python
new_df["Size_en"] = oe.transform(new_df[["Size"]] )
new_df
```

**ORDINAL ENCODING USING MAP FUN**

```python
ord_data1 = { "s":5 , "m":6 , "l":7 , "xl":10 }
```

```python
new_df["Size_en.map"] = new_df["Size"].map(ord_data1)
new_df
```
