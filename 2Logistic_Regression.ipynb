```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv("placement2.csv")
dataset.drop(columns="Student_ID", inplace=True)
dataset.rename(columns={"IQ":"Score", "Placement":"placed" , "CGPA":"cgpa"} , inplace=True)
dataset.head(10)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x = "cgpa" , y="Score" , data = dataset , hue="placed")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["placed"]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train , y_train)
```

```python
lr.score(x_test , y_test)*100
```

```python
lr.predict([[6.8 , 123]])
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy() , y.to_numpy() ,clf=lr)
plt.show()
```
