```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
from sklearn.tree import plot_tree
```

```python
dataset = pd.read_csv("/Users/bhavyakansal/AI/Datasets/Social_Network_Ads.csv")
```

```python
dataset.head(10)
```

```python
dataset.isnull().sum()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Purchased"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
sc.fit(x)
```

```python
x = pd.DataFrame(sc.transform(x), columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2 , random_state=42)
```

# Pre Pruning

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(max_depth=3)
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)*100
```

```python
dt.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(50,50))
plot_tree(dt)
plt.show()
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy() ,clf=dt)
plt.show()
```

# Post Pruning

```python
for i in range (1,20):
    dt1 = DecisionTreeClassifier(max_depth=i)
    dt1.fit(x_train,y_train)
    print(dt1.score(x_train,y_train)*100 , dt1.score(x_test,y_test)*100 , i)
```

```python
dt1 = DecisionTreeClassifier(max_depth=3)
dt1.fit(x_train,y_train)
```

```python
dt1.score(x_train,y_train)*100
```

```python
dt1.score(x_test,y_test)*100
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy() ,clf=dt1)
plt.show()
```
