# Import All the Necessary Libraries we need 

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import confusion_matrix
import joblib
```

# Import Dataset and see it

```python
dataset = pd.read_csv('/Users/bhavyakansal/Datasets/House Prices.csv')
dataset.head(10)
```

```python
dataset.isnull().sum()
```

# Lets drop unecessary columns

```python
dataset.drop(columns=["ID" , "Date"] , inplace = True)
```

```python
dataset.dtypes
```

# Now lets spilt input and output values

```python
x = dataset.iloc[:,1:]
y = dataset.iloc[:,0]
```

# Now Let's Split them into Training and Testing data 

```python
x_train, x_test , y_train , y_test = train_test_split( x , y , test_size=0.2 , random_state=42)
```

# Now lets try Models on it 

```python
lir = LinearRegression()
lir.fit(x_train , y_train)
```

```python
lir.score(x_train , y_train)*100 , lir.score(x_test , y_test)*100
```

```python
dt = DecisionTreeRegressor(max_depth=4)
dt.fit(x_train,y_train)
```

```python
dt.score(x_train , y_train)*100 , dt.score(x_test , y_test)*100
```

# Now lets save the model using Joblib

```python
joblib.dump(lir, 'house_price_model.pkl')
```

```python
joblib.dump(dt, 'house_price_model_dt.pkl')
```
