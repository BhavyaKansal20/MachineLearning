```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/lovnishverma/datasets/refs/heads/main/dataset-car-price.csv")
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Mileage'] = df['Mileage'].str.replace(' kmpl', '').str.replace(' km/kg', '').astype(float)

df['Engine'] = df['Engine'].str.replace(' CC', '').astype(float)

df['Power'] = df['Power'].astype(str).str.replace(' bhp', '').replace('null', np.nan).astype(float)
```

```python
df.drop(['Unnamed: 0', 'Name', 'Location', 'New_Price'], axis=1, inplace=True)
```

```python
df['Mileage'] = df['Mileage'].fillna(df['Mileage'].median())
df['Engine'] = df['Engine'].fillna(df['Engine'].median())
df['Power'] = df['Power'].fillna(df['Power'].median())
df['Seats'] = df['Seats'].fillna(df['Seats'].median())
```

```python
df = pd.get_dummies(df, columns=['Fuel_Type', 'Transmission', 'Owner_Type'], drop_first=True)
```

```python
df.head()
```

```python
df.info()
```

```python
x = df.drop("Price" , axis=1)
y = df["Price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y , test_size=0.2 , random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train , y_train)
```

```python
lr.score(x_train , y_train)*100 , lr.score(x_test,y_test)*100
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dr = DecisionTreeRegressor()
dr.fit(x_train , y_train)
```

```python
dr.score(x_train , y_train)*100 , dr.score(x_test,y_test)*100
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)
```

```python
rf.score(x_train , y_train)*100 , rf.score(x_test,y_test)*100
```
