```python
dataset = dataset.copy()

dataset = dataset.dropna(subset=["message_text"])
dataset = dataset[dataset["message_text"].str.strip() != ""]
```

```python
x = dataset.iloc[:,:-1]
y = dataset["class_label"]
```

```python
vectorizer = TfidfVectorizer(max_features=5000)
x_converted = vectorizer.fit_transform(x)
```

```python
models = {
    "LR": LogisticRegression(),
    "NB": MultinomialNB(),
}
```

```python
for name, model in models.items():
    score = cross_val_score(model, x_converted, y, cv=5).mean()
    print(name, score)
```
