# It is used to group the unlabeled datasets into a cluster and also known as hierarchical cluster analysis or HCA.
# In this algorithm, we develop the hierarchy of clusters in the form of a tree, and this tree-shaped structure is known as the dendrogram.

# Hierarchical Clustering Have two approaches i.e : 
# Agglomerative: Agglomerative is a bottom-up approach, in which the algorithm starts with taking all data points as single clusters and merging them until one cluster is left.
# Divisive: Divisive algorithm is the reverse of the agglomerative algorithm top-down approach.

# HOW TO MEASURE THE DISTANCE BETWEEN TWO CLUSTERS :
# The closest distance between the two clusters is crucial for the hierarchical clustering. There are various ways to calculate the distance between two clusters, and these ways decide the rule for clustering. These measures are called Linkage methods.
# Types : Single Linkage , Complete Linkage , Average Linkage , Centroid Linkage

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Lets Load Dataset and see it 

```python
dataset = pd.read_csv("/Users/bhavyakansal/Datasets/unlabeled_iris.csv")
dataset.head(12)
```

# Lets see that our dataset is Linear separable or not because Hierarchical Clustering works best on Linear datasets.

```python
sns.pairplot(data=dataset)
plt.show()
```

# Now let's first we will make a Dendogram of the dataset and find out the number of clusters in the dataset and then we will put that number of clusters into Agglomerative Clustering and then it tell us whcih groups are made and how . 

# So For Dendogram We have to use Scipy Library so lets install and use it .

```python
pip install Scipy
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,20))
sc.dendrogram(sc.linkage(dataset , method="single" , metric='euclidean'))
plt.savefig("Dendogram.jpg")
plt.show()
```

# Now lets find out the number of Clusters by finding most long line without cutting 

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=2 , linkage="single")
dataset["Predict"]= ac.fit_predict(dataset)
```

# Now lets Put and see the pair plot of that new dataset with predict column

```python
sns.pairplot(data = dataset , hue="Predict")
plt.show()
```
