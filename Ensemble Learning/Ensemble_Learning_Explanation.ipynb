#### * The Ensemble Learning/methods in machine learning combine the insights obtained from multiple learning models to facilitate accurate and improved decisions.

#### There are two main types of ensemble learning:
#### 1.Bagging (Bootstrap Aggregating) : Models are trained independently on different random subsets of the training data. Their results are then combined—usually by averaging (for regression) or voting (for classification). This helps reduce variance and prevents overfitting.

#### 2.Boosting : Models are trained one after another. Each new model focuses on fixing the errors made by the previous ones. The final prediction is a weighted combination of all models, which helps reduce bias and improve accuracy.

#### * Popular Types of Bagging Algorithms : 
#### Random Forest: The most popular bagging algorithm. It builds multiple decision trees on different data samples and random feature subsets to create a robust, uncorrelated "forest".
#### Extra Trees (Extremely Randomized Trees): Similar to Random Forest but adds more randomness by choosing split points at random instead of looking for the best possible split, which often makes it faster to train.
#### Bagged Decision Trees: A standard bagging approach where multiple deep decision trees are trained on bootstrapped samples and their results are averaged.
#### Pasting: A variant of bagging where data subsets are sampled without replacement, meaning each data point appears only once across subsets.
#### Random Subspaces: Instead of sampling data rows, this method samples features, training each model on a random subset of input features. 

#### * Popular Types of Boosting Algorithms : 
#### AdaBoost (Adaptive Boosting): One of the earliest boosting methods. It increases the weights of misclassified data points after each round, forcing the next model to focus more on difficult cases.
#### Gradient Boosting Machines (GBM): Trains new models on the residual errors (residuals) of previous models using gradient descent to minimize a loss function.
#### XGBoost (Extreme Gradient Boosting): An optimized and highly efficient implementation of gradient boosting known for its speed and performance. It includes built-in regularization to prevent overfitting.
#### LightGBM (Light Gradient Boosting Machine): Developed by Microsoft, this algorithm is optimized for large datasets and is often faster than XGBoost because it uses histogram-based learning.
#### CatBoost (Categorical Boosting): Specifically designed to handle categorical data natively without extensive preprocessing like one-hot encoding
